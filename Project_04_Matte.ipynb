```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Part 1: Loading the Dataset; Preliminary Analysis
In this section, we will load the data into a DataFrame, and will explore the structure of the data set.

```python
diamonds=pd.read_csv('diamonds.txt', sep='\t')
diamonds.head(10)
```

We will now determiine the size of the dataset.

```python
print(diamonds.shape)
```

We will now inspect the distribution of the columns in diamonds.

```python
diamonds.describe()
```

## Part 2: Filtering and Sorting
We will start by viewing information about the 5 most expensive diamonds in the dataset.

```python
diamonds.loc[:,['price','carat','color','clarity']].sort_values(by='carat',ascending=False).head(5)
```

```python
diamonds.loc[:,['price','carat','color','clarity']].sort_values(by='price').head(5)
```

We will now view information about the 5 largest diamonds in the dataset with an ideal cut.


```python
diamonds.loc[diamonds['cut']=='Ideal',['price','carat','cut','color','clarity']].sort_values(by='carat',ascending=False).head(5)
```

```python
diamonds.loc[diamonds['cut']=='Fair',['price','carat','cut','color','clarity']].sort_values(by='carat',ascending=False).head(5)
```

## Part 3: Working with Categorical Variables
We will now specify the order of the three categorical variables.

```python
clarity_levels=['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
cut_levels=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_levels=['J', 'I', 'H', 'G', 'F', 'E', 'D']
```

We will now use these lists to communicate to Pandas the correct order for the levels of the three categorical variables.

```python
diamonds.loc[:,'clarity']=pd.Categorical(diamonds.loc[:,'clarity'],clarity_levels)
diamonds.loc[:,'cut']=pd.Categorical(diamonds.loc[:,'cut'],cut_levels)
diamonds.loc[:,'color']=pd.Categorical(diamonds.loc[:,'color'],color_levels)
```

We will now create lists of named colors to serve as palettes to be used for visualizations later in the notebook.

```python
colors=['r','b','g','black','purple','brown','m','y','c']
clarity_pal=colors[:len(clarity_levels)]
cut_pal=colors[:len(cut_levels)]
color_pal=colors[:len(color_levels)]
```

## Part 4: Displaying Counts for Categorical Variables
We will start by counting the number of diamonds for each level of cut.

```python
diamonds.loc[:,'cut'].value_counts().sort_index()
```

We will now count the number of diamonds for each level of color.

```python
diamonds.loc[:,'color'].value_counts().sort_index()
```

We will now count the number of diamonds for each level of clarity.

```python
diamonds.loc[:,'clarity'].value_counts().sort_index()
```

## Part 5: Scatterplots of Price Against Carat
We will explore the relationship between the price and carat attributes by generating scatter plots.

```python
plt.figure(figsize=[8,6])
plt.scatter(diamonds.loc[:,'carat'],diamonds.loc[:,'price'],c='g',s=20,alpha=0.4)
plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title('Relationship between Diamond Price and Carat Size')
plt.show()
```

We will reproduce the plot from above, but will now color the points according to
their clarity.

```python
plt.figure(figsize=[8,6])
for i in range(len(clarity_levels)):
    plt.scatter(diamonds.loc[diamonds['clarity']==clarity_levels[i],'carat'],
                diamonds.loc[diamonds['clarity']==clarity_levels[i],'price'],
                c=clarity_pal[i],label=clarity_levels[i],s=20,alpha=0.4)
    plt.xlabel('Carat Size')
    plt.ylabel('Price (in $)')
    plt.title('Relationship between Diamond Price and Carat Size')
    plt.legend()
plt.show()
```

We will now separate the points for each of the eight clarity levels into its own subplot.

```python
plt.figure(figsize=[12,6])
for i in range(len(clarity_levels)):
    plt.subplot(2,4,i+1)
    plt.scatter(diamonds.loc[diamonds['clarity']==clarity_levels[i],'carat'],
                diamonds.loc[diamonds['clarity']==clarity_levels[i],'price'],
                c=clarity_pal[i],s=20,alpha=0.4)
    plt.xlabel('Carat Size')
    plt.ylabel('Price (in $)')
    plt.title(clarity_levels[i])
    plt.xlim([0,4])
    plt.ylim([0,20000])
plt.tight_layout()
plt.show()
```

## Part 6: Applying Logarithmic Transformations
We will start by adding two new columns to the diamonds in order to store the transformed variables.


```python
diamonds['ln_carat']=np.log(diamonds.loc[:,'carat'])
diamonds['ln_price']=np.log(diamonds.loc[:,'price'])
```

We will now use histograms to explore the distribution of the diamond prices, and the log of the diamond prices.

```python
plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.hist(diamonds['price'],bins=20,color='r',edgecolor='k')
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Prices')
plt.subplot(1,2,2)
plt.hist(diamonds['ln_price'],bins=20,color='b',edgecolor='k')
plt.xlabel('Natural Log of Price')
plt.ylabel('Count')
plt.title('Histogram of Log-Prices')
plt.show()
```

We will now use histograms to explore the distribution of the carat sizes, and the
log of the carat sizes.

```python
plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.hist(diamonds['carat'],bins=20,color='r',edgecolor='k')
plt.xlabel('Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Carat Sizes')
plt.subplot(1,2,2)
plt.hist(diamonds['ln_carat'],bins=20,color='b',edgecolor='k')
plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Log-Carat Sizes')
plt.show()
```

## Part 7: Scatterplot of Transformed Variables
We will create a scatterplot of ln_price against ln_carat, with points colored according to their clarity.

```python
plt.figure(figsize=[8,6])
for i in range(len(clarity_levels)):
    plt.scatter(diamonds.loc[diamonds['clarity']==clarity_levels[i],'ln_carat'],
                diamonds.loc[diamonds['clarity']==clarity_levels[i],'ln_price'],
                c=clarity_pal[i],label=clarity_levels[i],s=20,alpha=0.4)
    plt.xlabel('Natural Log of Carat Size')
    plt.ylabel('Natural Log of Price')
    plt.title('Relationship between Log-Price and Log-Carat Size')
    plt.legend()
plt.show()
```

## Part 8: Grouping by Cut
We will create a grouped DataFrame displaying the mean price and carat size for each cut level.

```python
gb_cut=diamonds.loc[:,['cut','price','carat']].groupby('cut').mean()           
gb_cut
```

```python
plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.bar(gb_cut.index,gb_cut.iloc[:,0],color=cut_pal,edgecolor='k')
plt.xlabel('Cut')
plt.ylabel('Mean Price')
plt.title('Mean Price by Cut')
plt.subplot(1,2,2)
plt.bar(gb_cut.index,gb_cut.iloc[:,1],color=cut_pal,edgecolor='k')
plt.xlabel('Cut')
plt.ylabel('Mean Carat Size')
plt.title('Mean Carat Size by Cut')
plt.show()
```

## Part 9: Grouping by Color
We will create a grouped DataFrame displaying the mean price and carat size for each color level.

```python
gb_color=diamonds.loc[:,['color','price','carat']].groupby('color').mean()
gb_color
```

We will now use bar charts to graphically display the information from the
DataFrame above.

```python
plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.bar(gb_color.index,gb_color.iloc[:,0],color=color_pal,edgecolor='k')
plt.xlabel('Color')
plt.ylabel('Mean Price')
plt.title('Mean Price by Color')
plt.subplot(1,2,2)
plt.bar(gb_color.index,gb_color.iloc[:,1],color=color_pal,edgecolor='k')
plt.xlabel('Color')
plt.ylabel('Mean Carat Size')
plt.title('Mean Carat Size by Color')
plt.show()
```

We will create a grouped DataFrame displaying the mean price and carat size for each clarity level.
